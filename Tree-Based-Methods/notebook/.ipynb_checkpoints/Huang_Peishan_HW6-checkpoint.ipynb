```python
df1=pd.read_csv('../data/aps_failure_training_set.csv',skiprows=20,na_values='na')
df1['label']=df1['class'].values
df1=df1.drop(labels='class',axis=1)

df2=pd.read_csv('../data/aps_failure_test_set.csv',skiprows=20,na_values='na')
df2['label']=df2['class'].values
df2=df2.drop(labels='class',axis=1)

df=pd.concat([df1,df2],ignore_index=True)
df
```

(c) Train a random forest to classify the data set. Do NOT compensate for class imbalance in the data set. Calculate the confusion matrix, ROC, AUC, and misclassification for training and test sets and report them (You may use pROC package). Calculate Out of Bag error estimate for your random forset and compare it to the test error.

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
X=X.fillna(X.median())

X_train=X.iloc[:60000,:]
y_train=y.iloc[:60000]
X_test=X.iloc[60000:,:].reset_index(drop=True)
y_test=y.iloc[60000:].reset_index(drop=True)

model=RandomForestClassifier(oob_score=True).fit(X_train,y_train)
y_pred=model.predict(X_test)

confusion_matrix=confusion_matrix(y_test,y_pred)
print(confusion_matrix)

proba=model.predict_proba(X_test)
pos_proba=proba[:,1]
roc_auc=roc_auc_score(y_test,pos_proba)
print(roc_auc)
```

```python
accuracy=accuracy_score(y_test,y_pred)
misclassification=1-accuracy
oob_score=model.oob_score_
oob_error=1-oob_score

print('The misclassification rate ',misclassification)
print('The out of bag error ', oob_error)
```

(e) XGBoost and Model Trees

```python
#uncompensate imbalanced dataset
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
X=X.fillna(X.median())

X_train=X.iloc[:60000,:].values
y_train=y.iloc[:60000]
y_train=y_train.replace(('neg','pos'),(0,1))
#print(y_train.value_counts())

alphas=[1e-4,1e-3,1e-2,1e-1,1e-0,1e+1,1e+2,1e+3,1e+4]

Accuracy=[]
k=0
tmp_i=0
for i in alphas:
    kfold=KFold(n_splits=5)
    average_cross_val=[]
    for train_index, test_index in kfold.split(X_train):
        X_tr, X_te=X_train[train_index],X_train[test_index]
        y_tr, y_te=y_train[train_index],y_train[test_index]
        model=XGBClassifier(reg_alpha=i,objective='binary:logistic').fit(X_tr,y_tr)       
        y_pred=model.predict(X_te)
        accuracy=accuracy_score(y_te,y_pred)
        average_cross_val.append(accuracy)
    Accuracy.append(np.mean(average_cross_val))
    if np.mean(average_cross_val)>k:
        k=np.mean(average_cross_val)
        tmp_i=i

misclassification=1-k
print(Accuracy)
print(f'The highest accuracy score is {k}, with alpha equal to {tmp_i}.')
```

```python
print('The highest average 5 fold cross validation accuracy score, ',k)
print('The lowest misclassification trainset error, ',misclassification)

X=df.iloc[:,:-1]
y=df.iloc[:,-1]
X=X.fillna(X.median())

X_train=X.iloc[:60000,:].values
y_train=y.iloc[:60000]
y_train=y_train.replace(('neg','pos'),(0,1))
X_test=X.iloc[60000:,:].reset_index(drop=True).values
y_test=y.iloc[60000:].reset_index(drop=True)
y_test=y_test.replace(('neg','pos'),(0,1))

model=XGBClassifier(reg_alpha=tmp_i,objective='binary:logistic').fit(X_train,y_train)
y_pred=model.predict(X_test)
accuracy=accuracy_score(y_test,y_pred)
misclassification=1-accuracy
print('The misclassification testset error ',misclassification)

confusion_matrix=confusion_matrix(y_test,y_pred)
print(confusion_matrix)

proba=model.predict_proba(X_test)
pos_proba=proba[:,1]
roc_auc=roc_auc_score(y_test,pos_proba)
print(roc_auc)
```

(f) Use SMOTE (Synthetic Minority Over-sampling Technique) to pre-process your data to compensate for class imbalance. Train XGBosst with L1-penalized logistic regression at each node using the pre-processed data and repeat 1e. Do not forget that there is a right and a wrong way of cross validation here. Compare the uncompensated case with SMOTE case.

```python
#compensate imbalanced dataset using SMOTE
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
X=X.fillna(X.median())
#only apply SMOTE on the training data
X_train=X.iloc[:60000,:]
y_train=y.iloc[:60000]

oversample=SMOTE()
X_train, y_train=oversample.fit_resample(X_train,y_train)
print(y_train.value_counts())
```

```python
y_train=y_train.replace(('neg','pos'),(0,1))

alphas=[1e-4,1e-3,1e-2,1e-1,1e-0,1e+1,1e+2,1e+3,1e+4]
Accuracy=[]
k=0
tmp_i=0
for i in alphas:
    average_cross_val=[]
    kfold=KFold(n_splits=5)
    for train_index, test_index in kfold.split(X_train):
        X_tr, X_te=X_train.iloc[train_index], X_train.iloc[test_index]
        y_tr, y_te=y_train.iloc[train_index], y_train.iloc[test_index]
        model=XGBClassifier(reg_alpha=i,objective='binary:logistic').fit(X_tr,y_tr)
        y_pred=model.predict(X_te)
        accuracy=accuracy_score(y_te,y_pred)
        average_cross_val.append(accuracy)
    Accuracy.append(np.mean(average_cross_val))
    if np.mean(average_cross_val)>k:
        k=np.mean(average_cross_val)
        tmp_i=i

misclassification=1-k
print(Accuracy)
print('The lowest missclassfication training error ', misclassification)
print('The selected alpha ', tmp_i)
```
