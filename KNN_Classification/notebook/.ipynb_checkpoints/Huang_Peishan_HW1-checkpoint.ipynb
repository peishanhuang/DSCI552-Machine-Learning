```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.neighbors import DistanceMetric
```

Question(b)i:

```python
#using pandas to read the file
data=pd.read_table('../data/column_2C.dat',sep=' ',header=None,names=('pelvic_incidence','pelvic_tilt','lumbar_lordosis_angle','sacral_slope','pelvic_radius','degree_SL','label'))
df=pd.DataFrame(data)
#to better display Dataframe 
pd.set_option('display.max_rows',500)
pd.set_option('display.max_columns',500)
pd.set_option('display.width',1000)
#convert labels to 0 and 1
df['label']=df['label'].replace(['AB','NO'],[1,0])
#use seaborn package to plot pairwise relationship between variables
sns.pairplot(df,hue='label')

```

Question(b)ii:

```python
fig, axes=plt.subplots(2,3,figsize=(18,10))
fig.suptitle("six independent variables boxplots")

sns.boxplot(ax=axes[0,0],data=df,x=df['pelvic_incidence'],y=df['label'])
sns.boxplot(ax=axes[0,1],data=df,x=df['pelvic_tilt'],y=df['label'])
sns.boxplot(ax=axes[0,2],data=df,x=df['lumbar_lordosis_angle'],y=df['label'])
sns.boxplot(ax=axes[1,0],data=df,x=df['sacral_slope'],y=df['label'])
sns.boxplot(ax=axes[1,1],data=df,x=df['pelvic_radius'],y=df['label'])
sns.boxplot(ax=axes[1,2],data=df,x=df['degree_SL'],y=df['label'])
```

Question(b)iii:

```python
#sSelect the first 70 rows of Class 0 and the first 140 rows of Class 1 as the training set
label_0=df[(df['label']==0)].iloc[:70,:]
label_1=df[(df['label']==1)].iloc[:140,:]
label_re0=df[(df['label']==0)].iloc[70:,:]
label_re1=df[(df['label']==1)].iloc[140:,:]

#Concatenating DataFrames and reset index
frame_tr=[label_0,label_1]
frame_te=[label_re0,label_re1]
trainset=pd.concat(frame_tr,ignore_index=True)
testset=pd.concat(frame_te,ignore_index=True)
```

Question(c)ii:

```python
#train test split
X_train=trainset.iloc[:,:-1].values
y_train=trainset.iloc[:,-1].values
X_test=testset.iloc[:,:-1].values
y_test=testset.iloc[:,-1].values
#create a set of k values and an empty list for train and test error
k_values=range(208,0,-3)
train_Errors=[]
test_Errors=[]
best_k=0
Min=2
#create the model
classifier=KNeighborsClassifier(p=2,weights='uniform')
#create a for-loop of different k for model
for k in k_values:
    classifier.n_neighbors=k
    classifier.fit(X_train,y_train)
    train_accuracy=classifier.score(X_train,y_train)
    train_Error=1-train_accuracy
    train_Errors.append(train_Error)
    test_accuracy=classifier.score(X_test,y_test)
    test_Error=1-test_accuracy
    if test_Error <=Min:
        Min = test_Error
        best_k = k
    test_Errors.append(test_Error)
print('the best k with minimum test error is:',(best_k,Min))
#print(len(train_Errors))

#plot the train and test error
plt.figure(figsize=(12,6))
plt.plot(k_values,train_Errors,color='red',label='Training Error',marker='.',linestyle='dashed')
plt.plot(k_values,test_Errors, color="blue",label='Testing Error',linestyle='dashed',marker='.')
plt.xlabel('K Values')
plt.ylabel('Mean Error')
plt.title('Performance under different K Values')
plt.legend(handlelength=4)
plt.show()
```

From the above figure, we can see when k=4, the model got the lowest mean testing error. Hence choose k=4.

```python
# choosing k=4 for creating classifier
classifier=KNeighborsClassifier(n_neighbors=4,p=2,weights='uniform')
classifier.fit(X_train,y_train)
y_pred=classifier.predict(X_test)
#evaluating model
matrix1=confusion_matrix(y_test,y_pred,labels=[1,0])
print('Confusion matrix:\n',matrix1)
tp, fn, fp, tn = matrix1.reshape(-1)
print('Outcome values:\n', tp,fn,fp,tn)
matrix2=classification_report(y_test,y_pred,labels=[1,0])
print('Classification report\n',matrix2)
print('true positive rate:',(69/(69+1)))
print('true negative rate:',(25/(25+5)))
```

Question(c)iii:

```python
#train and test set from (b)iii
trainset=pd.concat(frame_tr,ignore_index=True)
testset=pd.concat(frame_te,ignore_index=True)
N=range(10,211,10)
classifier=KNeighborsClassifier()
test_errors=[]
k_value=[]
x=0
y=2
for n in N:
    sub_0=trainset[(trainset['label']==0)].iloc[:int(n/3),:]
    sub_1=trainset[(trainset['label']==1)].iloc[:int(n-n/3),:]
    join_tr=[sub_0,sub_1]
    sub_trainset=pd.concat(join_tr,ignore_index=True)
    X_train=sub_trainset.iloc[:,:-1].values
    y_train=sub_trainset.iloc[:,-1].values
    X_test=testset.iloc[:,:-1].values
    y_test=testset.iloc[:,-1].values
    k=range(1,n,5)
    for i in k:
        classifier.n_neighbors=i
        classifier.fit(X_train,y_train)
        test_Accuracy=classifier.score(X_test,y_test)
        test_error=1-test_Accuracy
#Search best k and calculate the best test error(minimum) for each training set(totally 21)      
        if test_error<=y:
            y=test_error
            x=i
            test_errors.append(y)
            k_value.append(x)

print(len(k_value))

plt.figure(figsize=(15,7))
plt.plot(k_value,test_errors,color='green',label='Test Error',marker='.',linestyle='dashed')
plt.xlabel('K values')
plt.ylabel('Test Error')
plt.title('The Learning Curve')
plt.legend()
plt.show()
    
```

```python
trainset=pd.concat(frame_tr,ignore_index=True)
testset=pd.concat(frame_te,ignore_index=True)
#test & train split
X_train=trainset.iloc[:,:-1].values
y_train=trainset.iloc[:,-1].values
X_test=testset.iloc[:,:-1].values
y_test=testset.iloc[:,-1].values
#create the classifier
classifier=KNeighborsClassifier(p=1,metric='minkowski')
k=range(1,200,5)
#variables for plotting table
Test_errors=[]
K_values=[]
x=0
y=2
#create a for-loop of different k for model
for i in k:
    classifier.n_neighbors=i
    classifier.fit(X_train,y_train)
    test_accuracy=classifier.score(X_test,y_test)
    test_error=1-test_accuracy
    Test_errors.append(test_error)
    K_values.append(i)
    if test_error<=y:
        y=test_error
        x=i

#find the best k        
print('the best k with minimum test error is:',(x,y))    

#create the table from pd.DataFrame
data={'k values':K_values,'test error':Test_errors}
to_df=pd.DataFrame(data)
fig, ax=plt.subplots(figsize=(12,10))
fig.patch.set_visible(False)
ax.axis('off')
ax.axis('tight')
table=ax.table(cellText=to_df.values,rowLabels=to_df.index,colLabels=to_df.columns,loc='center')
plt.title('the Test Error Summarization with Manhattan Distance')
plt.show()
```

```python
# Euclidean distance

trainset=pd.concat(frame_tr,ignore_index=True)
testset=pd.concat(frame_te,ignore_index=True)
#train test split
X_train=trainset.iloc[:,:-1].values
y_train=trainset.iloc[:,-1].values
X_test=testset.iloc[:,:-1].values
y_test=testset.iloc[:,-1].values
#variables for plotting
k_values=[]
test_errors=[]
y=2
x=0
#create the model
k=range(1,200,5)
classifier=KNeighborsClassifier(metric='minkowski',p=2,weights='distance')
for i in k:
    classifier.n_neighbors=i
    classifier.fit(X_train,y_train)
    test_accuracy=classifier.score(X_test,y_test)
    test_error=1-test_accuracy
    test_errors.append(test_error)
    k_values.append(i)
    if test_error<=y:
        y=test_error
        x=i

print('the best k with minimum test error is: ',(x,y))
#plotting the table
data={'k values':k_values,'test errors':test_errors}
to_df=pd.DataFrame(data)
fig, ax=plt.subplots(figsize=(12,10))
fig.patch.set_visible(False)
ax.axis('off')
ax.axis('tight')
table=ax.table(cellText=to_df.values,rowLabels=to_df.index,colLabels=to_df.columns, loc='center')
plt.title('the Test Error Summarization with Euclidean Distance')
plt.show()
```
