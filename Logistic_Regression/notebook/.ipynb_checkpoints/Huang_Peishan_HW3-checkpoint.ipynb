```python
std=df.std()
boot_statistic=[]
CIs=[]
for i in df.columns:
    each_column=df[i]
    values=each_column.values
    for i in range(100):
        boot_sample=random.choices(values,k=88)
        boot_std=stat.stdev(boot_sample)
        boot_statistic.append(boot_std)
        while len(boot_statistic)==100:
            boot_statistic.sort()
            CI=[boot_statistic[5],boot_statistic[94]]
            CIs.append(CI)
            boot_statistic=[]

s1=pd.Series(std).reset_index()
s2=pd.Series(CIs)

print('The estimated standard deviation of each time-domain features is:\n',s1)
print('\n')
print('The 90% bootstrap confident interval for the standard deviation of each features is:\n',s2)
```
